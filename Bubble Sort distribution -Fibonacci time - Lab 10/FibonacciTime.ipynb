## 2. Fibonacci Time
Grafique la el tiempo de la función recursiva de Fibonaacci en el intervalo [0,50] y el tiempo de la función iterativa de Fibonacci [0,5000] incluyendo $y=n$ para comparar los resultados.


Función recursiva de Fibonacci

```
def FibR(n):
  if (n==0):
    return 0
  elif (n==1):
    return 1
  return (FibR(n-1)+ FibR(n-2))
```

```
def FibROptimized(n):
  memo={}
  def Fib(n):
    if (n==0):
      return 0
    elif (n==1):
      return 1
    if (n-2) not in memo:
      memo[n-2] = Fib(n-2)
    if (n-1) not in memo:
      memo[n-1] = Fib(n-1)
    return (memo[n-1] + memo[n-2])
  return Fib(n)
```

Función iterativa de Fibonacci

```
def FibI(n):
  if (n==0):
    return 0
  a, b = 0, 1
  for i in range(1,n):
    a, b = b, a+b 
  return b
```

Función de obtención de números de Fibonacci para un intervalo dado

```
def calcFb (tfb, liminf, limsup): #recibe el tipo de funcion de fibonacci 0=Recursiva, 1=Recuriva memorizada y 2 = Iterativa
  yt= [] #Lista para la funcion y(t)
  if (tfb == 0):
    for i in range(liminf, limsup+1,1):
      ti = time.clock()
      vfibonacci= FibR(i)
      tf = time.clock()
      yt.append(tf-ti)
      #print(i)
    return (yt)
  elif (tfb == 1):
    for j in range(liminf, limsup+1,1):
      ti = time.clock()
      vfibonacci= FibROptimized(j)
      tf = time.clock()
      yt.append(tf-ti)
    return (yt)
  elif (tfb == 2):
    for k in range(liminf, limsup+1,1):
      ti = time.clock()
      vfibonacci= FibI(k)
      tf = time.clock()
      yt.append(tf-ti)
    return (yt)
  else: 
    print('Valor tfb incorrecto, debe estar entre 0 y 2')
    return (yt)
```

```
yn = list(range(0,51,1))
yni = list(range(0,501,1))
```

**Funcion Recursiva de Fibonacci [0,50]**

```
yt1 = calcFb(0, 0, 50)
p1 = plot(yt1)
title('Función Fibonacci Recursiva')
xlabel('N')
ylabel('Tiempo')
plt.show()
title('Función Fibonacci Recursiva Vs Orden líneal')
xlabel('N')
ylabel('Tiempo')
p3 = plot(yt1)
pni = plot(yn)
legend(('FbR', 'y=n'))
plt.show()
```

**Función Recursiva de Fibonacci Memorizada [0,50]**

```
yt2 = calcFb(1,0,50) 
p2 = plot(yt2)
title('Función Fibonacci Memorizada')
xlabel('N')
ylabel('Tiempo')
legend('FbRM')
plt.show()
title('Función Fibonacci Recursiva Memoria Vs Orden líneal')
xlabel('N')
ylabel('Tiempo')
p3 = plot(yt2)
pni = plot(yn)
legend(('FbRM', 'y=n'))
plt.show()
```

**Función Fibonacci Iterativa**

```
yt3 = calcFb(2,0,500)
p3 = plot(yt3)
title('Función Fibonacci Iterativa')
xlabel('N')
ylabel('Tiempo')
plt.show()
title('Función Fibonacci Iterativa Vs Orden líneal')
xlabel('N')
ylabel('Tiempo')
p3 = plot(yt3)
pni = plot(yni)
legend(('FbI', 'y=n'))
plt.show()
```
