## 1. Bubble Sort Distribution

Se inicia con la construcción del algoritmo de Bubble Sort obtenido de ungrid

```
def swap(v, i,j): #realiza el intercambio de posiciones
  v[i], v[j] = v[j], v[i]

def bubblesort(v): #Realiza el ordenamiento por el método de bubblesort
  pasos=0
  #swaps=0
  cadena = v.copy()
  for i in range(0, len(cadena), 1):
    for j in range(len(cadena)-1, i, -1):
      pasos+=1
      if (cadena[j]<cadena[j-1]):
        swap(cadena, j,j-1)
        pasos+=1
        #swaps+=1
  return pasos

def permutaciones(n):
  sortedlist = list(range(1,n+1)) #crea la lsita ordenada , range permite generar la lista a partir de un n dado
  permlist = list(itertools.permutations(sortedlist))
  steps = [] #guarda el número de pasos
  for i in permlist:
    st= bubblesort(list(i))
    steps.append(st)
  return steps
```

Ahora la función que graficará el número de pasos

```
def graficar_bar(datos,n): #función para hacer un histograma, trae como atributos los datos a graficar y el n que es el tamaño del arreglo
  titulo = 'Distribución de pasos de ordenamiento para un arreglo de tamaño ' + str(n)
  indices, valores = np.unique(datos, return_counts=True) #guarda los índices y valores del arreglo al ordenarlo por tipo de dato y su frecuencia
  fig = plt.figure()
  ax = fig.add_axes([0,0,1,1])
  ax.bar(indices,valores, color = 'r')
  ax.set_title(titulo)
  ax.set_ylabel('Frecuencia de Pasos')
  ax.set_xlabel('Número de Pasos')
  plt.show()
```

```
for k in range(0,12,1):
  pasos = permutaciones(k)
  graficar_bar(pasos, k)
```
